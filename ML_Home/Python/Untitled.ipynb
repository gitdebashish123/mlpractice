```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("/Users/debashish.kheti/Desktop/ML_Home/input_files/subs_counts.xlsx")

data['Rptg_Dt']=pd.to_datetime(data['Rptg_Dt'])
```

```python
data.index=data['Rptg_Dt']
data.drop(['Rptg_Dt'],axis=1,inplace=True)
data['count'] = np.cbrt(data['count'])
```

```python
data.head
```

```python
plt.figure(figsize=(10,4))
plt.plot(data)
plt.show()
```

```python
rollmean=data.rolling(2).mean()

```

```python
plt.figure(figsize=(10,4))
plt.plot(data)
plt.plot(rollmean,'r')
plt.show()
```

```python
datalog=np.log(data)
datalogd=datalog-datalog.shift(1)
plt.figure(figsize=(10,4))
plt.plot(datalogd)
plt.show()
```

```python
rollmean2=datalogd.rolling(4).mean()
plt.figure(figsize=(10,4))
plt.plot(datalogd)
plt.plot(rollmean2,'r')
plt.show()
```
