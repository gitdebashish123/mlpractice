<h1>Handling Missing Data</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/debashish.kheti/Desktop/ML_Home/input_files/nyc_weather.csv",parse_dates=['EST'])
df.head(10)
#type(df.EST[0])
df.set_index('EST',inplace=True)
df.head(5)
```

```python
#Method1 -set default
new_df =df.fillna(0)
new_df
#new_df.head(5)
```

```python
#Method1 -set dictionary
new_df=df.fillna({'Events':'No Events','WindSpeedMPH':0})
new_df.head(5)
```

```python
#forward or backward fill
new_df=df.fillna(method='bfill',limit=1)
new_df.head(5)
# There are other option along with ffill ,axis="columns" transfer the values horizontally 
#or limit=1 pass the value ony once
```

```python
#Interpolation
new_df=df.interpolate()
new_df.head(5)
new_df=df.interpolate(method="time")#also uses time as an factor to determine nearby values
```

```python
#Dropping all bad records
new_df=df.dropna()
new_df.head(5)
new_df.shape
```

```python
new_df=df.dropna(how="all") #deletes rows only when all values are null
new_df.head(5)
new_df.shape
new_df=df.dropna(thresh=1) #dont drop the value if it has at least one NaN
new_df.shape
```
