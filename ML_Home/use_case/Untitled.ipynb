```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from pandas import DataFrame
from pandas.tools.plotting import autocorrelation_plot
```

```python
data=pd.read_excel("/Users/debashish.kheti/Desktop/ML_Home/input_files/subscription_counts.xlsx")
data.head()
```

```python
data['rptg_dt']=pd.to_datetime(data['rptg_dt'])
```

```python
data.index=data['rptg_dt']
data.drop(['rptg_dt'],axis=1,inplace=True)
data.head()
```

```python
data.plot()
plt.show()
```

```python
autocorrelation_plot(data)
plt.show()
```

```python
rollmean=data.rolling(2).mean()
rollmean
```
