```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from pandas import DataFrame
from pandas.tools.plotting import autocorrelation_plot
```

```python
def parser(x):
	return datetime.datetime.strptime('19'+x,'%Y-%m')
```

```python
series=pd.read_csv("/Users/debashish.kheti/Desktop/ML_Home/input_files/sales-of-shampoo-over-a-three-ye.csv",header=0,nrows=36,parse_dates=[0],index_col=0,squeeze=True,date_parser=parser)
series.head()

```

```python
series.plot()
plt.show()
```

```python
autocorrelation_plot(series)
plt.show()
```

```python
rollmean=series.rolling(5).mean()
rollmean
```

```python
plt.figure(figsize=(10,4))
plt.plot(series)
plt.plot(rollmean,'r')
plt.show()
#if rollmean is straight,there is no trend
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(series,(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
residuals = DataFrame(model_fit.resid)
residuals.plot()
plt.show()
```
